Token level embedding


```python
from transformers import AutoTokenizer, AutoModel
```

```python
tokenizer= AutoTokenizer.from_pretrained("bert-base-uncased")
model= AutoModel.from_pretrained("bert-base-uncased")
```

```python
#Tokenizing the sentence
tokens=tokenizer("Great learning embeddings",return_tensors='pt')
```

```python
tokens
```

```python
#Creating contextual embeddings
embeddings=model(**tokens)[0]
```

```python
embeddings
```

```python
embeddings.shape
```

**Sentence Level Embeddings**

```python
from sentence_transformers import SentenceTransformer

#Loading the model

model= SentenceTransformer('all-MiniLM-L6-v2')
```

```python
#Convert text to text embeddings
embeddings=model.encode("Great learning embeddings, Great learning")
```
